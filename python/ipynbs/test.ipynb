```python
import requests
import time


validator_request = requests.get('https://solanacompass.com/stats')
```

```python
validators = [i['key'] for i in validator_request.json()['validators_table']]
```

```python
validator_data = []
interested_validators = []
i = 0
while i < len(validators):
    request = requests.get(f'https://solanacompass.com/stats?validator={validators[i]}').json()
    validator_data.append(request)
    if request['validators_table'][0]['commission'] >10 and request['validators_table'][0]['wallets'] <300:
        i += 1
        continue
    request = request['epochCharts']['balance']['data']
    key = sorted(request.keys())
    max_stake = max(request.values())
    if max_stake > 20000 and request[key[-1]] < 0.2*max_stake:
        print([(key,value) for key,value in request.items() if value == max_stake], " -> ", request[key[-1]])
        print(validators[i])
        interested_validators.append(validator_data[-1])
    i += 1
    if not i%500:
        print("reached count: ", i)
        time.sleep(60)


```
